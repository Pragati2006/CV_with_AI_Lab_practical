```python
import pandas as pd

data = [
    ["Sunny", "Warm","High","Weak","Yes"],
    ["Sunny", "Warm","High","Strong","Yes"],
    ["Cloudy", "Warm","High","Weak","Yes"],
    ["Rainy", "Cold","High","Weak","No"],
    ["Sunny", "Warm","Normal","Weak","Yes"],
    ["Cloudy", "Cold","Normal","Strong","No"],
]
columns = ["Sky", " Termperature","Humidity","Wind","Rain"]
df = pd.DataFrame(data, columns = columns)

csv_path = "weather_rain.csv"
df.to_csv(csv_path, index=True)

df
```

```python
def find_s(csv_path: str, target_col: str= None, positive_label:str = "Yes"):
    df = pd.read_csv(csv_path)
    if target_col is None:
        target_col = df.columns[-1]
    attributes = [c for c in df.columns if c != target_col]
    x = df[attributes].values
    y = df[target_col].astype(str).values
        
    S=None
    steps=[]
    for i in range(len(x)):
        label = y[i].strip()
            
        if label == positive_label:
            if S is None:
                S = list(x[i])
            else:
                for j in range(len(S)):
                    if S[j] != x[i][j]:
                        S[j] = "?"
            steps.append((i+1, "Positive", list(x[i]), list(S)))
        else:
            steps.append((i+1,"Negative", list(x[i]), None))
    return df, attributes , steps, S
```

```python
df , attributes , steps, final_S = find_s("weather_rain.csv", target_col="Rain", positive_label="Yes")
print("Attributes:", attributes)
print("\n--- Step-wise Demonstration (Find-S) ---")
for idx, kind, instance, hyp in steps:
    if kind == "Positive":
        print(f"\nInstance {idx} (POSITIVE): {instance}")
        print(f"Updated S-hypothesis: {hyp}")
    else:
        print(f"\nInstance {idx}(NEGATIVE):{instance}")
        print("Ignored in Find-S.")
print("\n--- Final Output ---")
print("Most Specific Hypothesis (Final S):", final_S)
```
