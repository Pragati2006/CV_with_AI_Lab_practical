```python
import pandas as pd
import numpy as np
```

```python
raw_data = {
    "Roll_No":[101,102,103,104,104,105,106,107,108,109,110,111],
    "Name":["Aman","Bhavna","Chetan","Diya","Diya","Eshan","Farah","Gopal","Heena","Irfan","Jatin","Kiran"],
    "Gender":["Male","Female","Male","Female","Female","Male","Female","Male","Female","Male","Male","Female"],
    "Marks":[78,85,np.nan,92,92,999,66,73,88,np.nan,59,81],
    "Attendance":["85%","90%","80 percent","95%","95%","60%","72%","88%",None,"91%","77%","89%"],
    "Exam_Date":["2026-02-01","01/02/2026","2026.02.01","Feb 1 2026","Feb 1 2026","2026-02-01",
                "2026/02/01","2026-02-01","02-01-2026","2026-02-01","invalid_date","2026-02-01"]
}
csv_raw = pd.DataFrame(raw_data)
raw_csv_path="student_marks_raw.csv"
csv_raw.to_csv(raw_csv_path,index=False)
print("raw Dataset saved as:", raw_csv_path)
csv_raw
```

```python
df = pd.read_csv("student_marks_raw.csv")
df.head()
```

```python
print("Duplicates(count):",df.duplicated().sum())
print("Shape before removing duplicates:",df.shape)
df = df.drop_duplicates()
print("Shape after removing duplicates:",df.shape)
df
```

```python
df["Attendance"]=(
    df["Attendance"]
    .astype(str)
    .str.lower()
    .str.replace("%","",regex=False)
    .str.replace("percent","",regex=False)
    .str.strip()
)
df["Attendance"] = pd.to_numeric(df["Attendance"],errors="coerce")
print(df[["Attendance"]].head(10))
print("\nMissing in Attendance after conversion:",df["Attendance"].isnull().sum())
```

```python
df.loc[df["Marks"]>100,"Marks"]=np.nan
marks_mean = df["Marks"].mean(skipna=True)
attendance_median = df["Attendance"].median(skipna=True)
df["Marks"]=df["Marks"].fillna(round(marks_mean,2))
df["Attendance"]=df["Attendance"].fillna(attendance_median)
print("Marks mean:",marks_mean)
print("Attendance median:",attendance_median)
df.isnull().sum()
```

```python
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})
df[["Gender"]].head()
```

```python
print("Final shape:",df.shape)
print("\nMissing Values:\n",df.isnull().sum())
print("\nData types:\n",df.dtypes)
df.head(10)
```

```python
df["Exam_Date"]=pd.to_datetime(df["Exam_Date"],errors="coerce",dayfirst=True)
most_common_date = df["Exam_Date"].mode()[0]
df["Exam_Date"]=df["Exam_Date"].fillna(most_common_date)
df["Exam_Date"]=df["Exam_Date"].dt.strftime("%Y-%m-%d")
df["Exam_Date"].dtype
df
```
