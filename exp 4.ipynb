```python
df = pd.read_csv("countries-aggregated.csv")
print(df.shape)
print(df.head())
df["Active"]= df["Confirmed"] - df["Recovered"] - df["Deaths"]
```

```python
df["Date"]= pd.to_datetime(df["Date"])
df["Active"]= df["Confirmed"]-df["Recovered"]-df["Deaths"]
df["Active"]=df["Active"].clip(lower=0)
df.head()
```

```python
latest_date = df["Date"].max()
df_latest = df[df["Date"]==latest_date].copy()
df_latest.head()
```

```python
fig_map=px.choropleth(
    df_latest,
    locations = "Country",
    locationmode = "country names",
    color = "Active",
    hover_name="Country",
    color_continuous_scale = "Reds",
    title= f"COVID-19 Active Cases - {latest_date.date()}"
)
fig_map.show()
```

```python
top15 = df_latest.sort_values("Active",ascending=False).head(15)
fig_bar= px.bar(
    top15,
    x="Country",
    y = "Active",
    title="Top 15 Countries by Active COVID-19 Cases",
    text ="Active"
)
fig_bar.update_traces(texttemplate="%{text:,}",textposition="outside")
fig_bar.update_layout(xaxis_tickangle=-35)
fig_bar.show()
```

```python
world = df.groupby("Date")[["Confirmed","Recovered","Deaths","Active"]].sum().reset_index()
fig_line = px.line(
    world,
    x="Date",
    y="Active",
    title="Global Active COVID-19 cases over time"
)
fig_line.show()
```
